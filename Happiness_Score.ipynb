### Question no 1: What are the top 10 happiest and least happy countries?

### Question 2 : How does happiness score vary across regions?

```python
happiness_vary_across_region = df.groupby('region')['happiness_score'].mean().sort_values (ascending = True )
happiness_vary_across_region
```

```python
plt.figure(figsize = (8,4))
happiness_vary_across_region.plot(kind = 'barh', color = 'green')
plt.xlabel('Happiness Score')
plt.ylabel('Region')
plt.title('Happiness Across Region')
plt.show()
```

 This plot shows how happiness vary across different regions, with North America and ANZ at the top most while 
  south Asia is the least region with the happiness score


## Question no 3: Is there a correlation between GDP per capita and happiness score?


```python
corr_gdp_happiness = df[['gdp_per_capita', 'happiness_score']].corr().iloc[0, 1]
print(f" The correlation between gdp_per_capita and Happiness score is :{corr_gdp_happiness }")
```

```python
df.plot(x = "gdp_per_capita", y= "happiness_score",
kind ="scatter",
color ="blue")
plt.title('Correlation between GDP per capita and Happiness score')
plt.xlabel('gdp_per_capita')
plt.ylabel ('happiness_score')
plt.show()
```

 The correlation between gdp per capita and happiness score is 0.7843, which show a positive correlation 
 beween them. It show that there is a strong relationship between the two.

## Question no 5: How does healthy life expectancy impact happiness across regions?

## Question no 6:   Which regions have the highest perceptions of corruption?

```python
highest_perception_corruption = df.groupby('region')['perceptions_of_corruption'].mean().sort_values(ascending= False)
print("The region with the highest perceptions of corruptions are:")
print(highest_perception_corruption)
```

```python
plt.figure(figsize=(8, 4))
highest_perception_corruption.plot(kind='barh', color='blue')
plt.xlabel('Average Perceptions of Corruption')
plt.ylabel('Region')
plt.title('Top 10 Regions with the Highest Perceptions of Corruption')
plt.gca().invert_yaxis() 
plt.show()
```

 inference: The data reveals that perceptions of corruption vary significantly across regions. North America and ANZ (Australia and New Zealand) 
 exhibit the highest perception of corruption with a score of 0.329, indicating that these regions perceive corruption 
 more prominently compared to others. Western Europe follows closely with a score of 0.302, suggesting similar concerns 
 about corruption in this region.
 Regions with the lowest perceptions of corruption include Central and Eastern Europe (0.076), and Sub-Saharan Africa (0.103)
 indicating that individuals in these areas perceive corruption as less of an issue relative to the other regions listed.


## Question no 7:  Is there a link between generosity and happiness?

```python
link_generosity_happiness= df['generosity'].corr(df['happiness_score'])
link_generosity_happiness
```

```python
plt.figure(figsize = (8, 4))
sns.scatterplot(data=df, x = 'generosity', y = 'happiness_score')
plt.title('Link between Generosity and Happiness Score')
plt.xlabel('Generosity')
plt.ylabel('Happiness Score')
color= 'green'
plt.show()
```

 The score of 0.04367 shows that there is a weak relationship or correlation between Generosity and Happiness score. 
 It implies that generosity has very little or no impact on the happiness score. 

## Question no 8: How does the freedom to make life choices relate to happiness?

```python
freedom_choices_relate_happiness = df['freedom_to_make_life_choices'].corr(df['happiness_score'])
freedom_choices_relate_happiness                                                                          
```

```python
plt.figure(figsize = (8, 4))
sns.scatterplot(data=df, x= 'freedom_to_make_life_choices', y = 'happiness_score')
plt.xlabel('Freedom to Make Life Choices')
plt.ylabel('Happiness')
plt.title('The Relatioship between Life Choices and Happiness')
plt.show()
```

Inference: The correlation of 0.663 shows that there is a positive correlation or relationship between freedom to life choices and happiness. 
It simply imply that as freedom to life choices increases, happiness also increases.

## Question no 9: What are the averages for each metric across regions?

```python
numeric_columns = df.columns[2:] 
average_metric_across_region = df.groupby('region')[numeric_columns].mean()
print(average_metric_across_region)
```

 inference: Regions with higher GDP per capita, social support, and healthy life expectancy tend to have higher happiness scores, especially in 
 North America and Western Europe. In contrast, regions like South Asia and Sub-Saharan Africa with lower economic and social factors show lower 
 happiness scores. Generosity and perceptions of corruption also vary, influencing happiness differently across regions


## Question no 10: Which countries deviate the most from the regional average happiness score?

```python
country_deviate_happiness = df.groupby('region')['happiness_score'].mean().sort_values(ascending=False)
country_deviate_happiness
```

```python
plt.figure(figsize=(8, 4))
country_deviate_happiness.plot(kind='barh', color= 'blue')
plt.xlabel('Region')
plt.ylabel('Happiness Score')
plt.title('Countries that Deviate Most from the Regional Average Happiness Score')
plt.gca().invert_yaxis()
plt.show()
```

The country that deviate the most is North America and ANZ with th average happiness score of 7.018 while the least is south Asis with average happiness score of 4.089
